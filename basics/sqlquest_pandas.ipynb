```python
data = [[1, 1], [1, 2], [1, 3], [2, 1], [2, 4]]
project = pd.DataFrame(data, columns=['project_id', 'employee_id']).astype({'project_id':'Int64', 'employee_id':'Int64'})
data = [[1, 'Khaled', 3], [2, 'Ali', 2], [3, 'John', 1], [4, 'Doe', 2]]
employee = pd.DataFrame(data, columns=['employee_id', 'name', 'experience_years']).astype({'employee_id':'Int64', 'name':'object', 'experience_years':'Int64'})

project
```

```python
employee
```

```python
merged = pd.merge(project, employee, on = 'employee_id')
merged = merged.groupby('project_id')['experience_years'].mean().reset_index().rename(columns={'experience_years': 'average_years'})
merged['average_years'] = round(merged['average_years'], 2)
merged
```

```python
data = [[1, 23], [2, 9], [4, 30], [5, 54], [6, 96], [7, 54], [8, 54]]
visits = pd.DataFrame(data, columns=['visit_id', 'customer_id']).astype({'visit_id':'Int64', 'customer_id':'Int64'})
data = [[2, 5, 310], [3, 5, 300], [9, 5, 200], [12, 1, 910], [13, 2, 970]]
transactions = pd.DataFrame(data, columns=['transaction_id', 'visit_id', 'amount']).astype({'transaction_id':'Int64', 'visit_id':'Int64', 'amount':'Int64'})
```

```python
df1 = pd.DataFrame(visits)
df2 = pd.DataFrame(transactions)
newdf = df1.merge(df2,on='visit_id',how='left')
tranz=newdf[newdf['transaction_id'].isnull()]
result = (tranz.groupby('customer_id').count().sort_values('visit_id', ascending=False).rename(columns={'visit_id':'count_no_trans'})[['count_no_trans']]
)
result = result.reset_index()
result
```

```python
data = [[9, 'Hercy', None, 43], [6, 'Alice', 9, 41], [4, 'Bob', 9, 36], [2, 'Winston', None, 37]]
employees = pd.DataFrame(data, columns=['employee_id', 'name', 'reports_to', 'age']).astype({'employee_id':'Int64', 'name':'object', 'reports_to':'Int64', 'age':'Int64'})
```

```python
employees
```

```python
managers = employees[employees['employee_id'].isin(employees['reports_to'].dropna())]
manager_reports = employees.groupby('reports_to').agg(num_reports=('employee_id', 'count'),avg_age=('age', 'mean')).reset_index()
manager_details = pd.merge(managers[['employee_id', 'name']], manager_reports,
                               left_on='employee_id', right_on='reports_to',
                               how='inner')[['employee_id', 'name', 'num_reports', 'avg_age']]
manager_details['avg_age'] = manager_details['avg_age'].round().astype('Int64')
manager_details = manager_details.sort_values(by='employee_id').reset_index(drop=True)
manager_details
```
