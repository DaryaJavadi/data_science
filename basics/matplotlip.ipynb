```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hmavrodiev/london-bike-sharing-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/london_merged.csv")
df.head()
```

```python
df["timestamp"].dtype
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
df['timestamp'].dt.year
```

```python
w = df.groupby(['is_weekend'])['cnt'].mean()
plt.bar(['weekday', 'weekend'], w.values);
```

```python
y = df.groupby(['is_holiday'])['cnt'].mean()
y
```

```python
y = df.groupby(['is_holiday'])['cnt'].mean()
plt.bar(['not holiday', 'holiday'], y.values);
```

```python
z = df.groupby(df['timestamp'].dt.hour)['cnt'].mean().plot(kind = 'bar', figsize = (12, 8));
```

```python
z = df.groupby(df[df['is_weekend']==1]['timestamp'].dt.hour)['cnt'].mean().plot(kind = 'bar', figsize = (12, 8));
```

```python
z = df.groupby(df['timestamp'].dt.month)['cnt'].mean().plot(kind = 'bar', figsize = (12, 8));
```

```python
z = df.groupby(df['season'])['cnt'].mean()
plt.pie(x = z.values, labels = ['spring', 'summer', 'autumn', 'winter'], autopct = '%.2f%%');
```

```python
df.columns
```

```python
import seaborn as sns
c = df[['cnt', 't1', 'hum', 'wind_speed']].corr()
sns.heatmap(c, annot = True, cmap = 'coolwarm');
```

```python
plt.bar(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],df.groupby(df['timestamp'].dt.weekday)['cnt'].mean().values);
```

```python
df_filtered = df[(df['timestamp'] > "2015-01-01")&(df['timestamp'] < "2015-01-31")]
df_filtered = df_filtered.groupby(df_filtered['timestamp'].dt.day)['cnt'].mean().reset_index()
plt.figure(figsize = (10, 7))
plt.plot(df_filtered['timestamp'], df_filtered['cnt'].diff(7));
```
